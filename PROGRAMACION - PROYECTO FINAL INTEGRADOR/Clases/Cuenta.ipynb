```python
from colorama import init, Fore, Back, Style
import pyttsx3

class Cuenta:

    @staticmethod
    def menu_cuenta(cuenta):
        opcion = None

        print(Fore.BLUE+ "**************************************************")
        print("\n**   BIENVENIDO AL BANCO SUCURSAL 444 CÓRDOBA **\n")
        print(Fore.BLUE+ "**************************************************")

        while opcion != 5:
            print(Fore.BLUE+ "****************************************")
            print("*                                      *")
            print("*         MENU PRINCIPAL               *")
            print("*                                      *")
            print(Fore.BLUE+ "****************************************")
            print("""
            
              Digite una Opcion:
             ========================
            || 1 => Depositar       || 
             ========================
            || 2 => Extraer         ||
             ========================
            || 3 => Consultar saldo ||
             ========================
            || 4 => Plazo fijo      ||
             ========================
            || 5 => Salir           ||
             ========================
            """)
            
            opcion = int(input("Digite una Opcion del Menú: "))
            if opcion == 1:
                cuenta.deposito(float(input("Ingrese el Monto Depositar: $")))
                print("=================================")
                print("El Deposito se Realizo con Exito")
                print("=================================")
                
            elif opcion == 2:
                cuenta.extraccion(float(input("Ingrese el Monto a Extraer: $")))

            elif opcion == 3:
                print("=================================")
                print("Saldo Actual es:",cuenta.saldo())
                print("=================================")      

            elif opcion == 4:
                print("=============================================================================")
                cuenta.plazo_fijo(
                    float(input("Ingrese el Monto del Plazo Fijo: $")),
                    int(input("Ingrese los Dias para el Plazo Fijo con Tasa de Interés Variable: ")),
                    float(input("Ingrese la Tasa de Interes: "))
                    )
                print("=============================================================================")
            elif opcion == 5:
                break

    def __init__(self):
        self._saldo = 0

    def deposito(self, monto):
        self._saldo += monto

    def extraccion(self, monto):
        self._saldo -= monto

    def saldo(self):
        return self._saldo

    def plazo_fijo(self, capital, tiempo, tasa_interes):
        interes = (capital * tasa_interes) / 100
        print("==============================")
        print(f'====== Plazo Fijo ===========')
        print("==============================")
        print(f'Capital invertido ${capital}')
        print("==============================")
        print(f'Disponible en: {tiempo} dias')
        print("==============================")
        print(f'Ganancia de ${interes}')
        print("==============================")



if __name__ == '__main__':
    cuenta = Cuenta()
    cuenta.menu_cuenta(cuenta)
```
